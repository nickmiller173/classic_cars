```python
import json
import pandas as pd

# Load data
data = pd.read_csv("../data/full_training_data.csv")
data.head()
# Get unique values for every categorical column
options = {
    "Make": sorted(data['Make'].unique().tolist()),
    "Model": sorted(data['Model'].unique().tolist()),
    "Exterior Color": sorted(data['Exterior Color'].unique().tolist()),
    "Interior Color": sorted(data['Interior Color'].unique().tolist()),
    "Title Status": sorted(data['Title Status'].unique().tolist()),
    "Seller Type": sorted(data['Seller Type'].unique().tolist()),
    "Drivetrain": sorted(data['Drivetrain'].unique().tolist()),
    "Body Style": sorted(data['Body Style'].unique().tolist()),
    "Transmission_Type": sorted(data['Transmission_Type'].unique().tolist()),
    "Engine_Cylinders": sorted(data['Engine_Cylinders'].unique().tolist())
}

# Save to a JSON file
with open('app_options.json', 'w') as f:
    json.dump(options, f)
```
