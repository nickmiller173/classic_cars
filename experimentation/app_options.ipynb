```python
import pandas as pd

def check_distributions(file_path):
    # Load the dataset
    df = pd.read_csv(file_path)

    print(f"--- Dataset Summary: {len(df)} rows, {len(df.columns)} columns ---\n")

    for col in df.columns:
        print(f"### Column: {col} ###")
        
        # 1. Print Value Counts (Top 10 for brevity)
        print("Top 10 Value Counts:")
        print(df[col].value_counts().head(10))
        
        # 2. Check Distribution/Summary Stats
        if pd.api.types.is_numeric_dtype(df[col]):
            # For numbers: Mean, Median, Min, Max
            print("\nNumeric Distribution:")
            print(df[col].describe().to_string())
        else:
            # For categories: Unique count and mode
            print("\nCategorical Distribution:")
            print(f"Unique values: {df[col].nunique()}")
            print(f"Most frequent: {df[col].mode()[0] if not df[col].mode().empty else 'N/A'}")
            
        print("-" * 30 + "\n")

# Usage
check_distributions("../data/train_data.csv")
```

```python
import pandas as pd

def get_main_color(x):
    # 1. Catch missing values and group them into "Other"
    if pd.isna(x):
        return "Other"
        
    # 2. Grab the primary color before slashes or " and "
    x = str(x).split('/')[0].split(' and ')[0].strip()
    x_lower = x.lower()
    
    # 3. Check for specific keywords FIRST to prevent substring collisions 
    special_map = {
        # Edge cases, Collisions & Exterior Bleed-over
        'titanium': 'Gray', 'titan': 'Black', 'mustang': 'Brown', 'tanzanite': 'Blue',
        'stainless': 'Silver', 'mercury': 'Silver', 'magnetic': 'Gray', 'thunder': 'Gray',
        
        # Blacks / Darks
        'ebony': 'Black', 'nero': 'Black', 'carbon': 'Black', 'onyx': 'Black', 
        'jet': 'Black', 'obsidian': 'Black', 'beluga': 'Black', 'panther': 'Black',
        'amido': 'Black', 'midnight': 'Black', 'anthracite': 'Gray', 'zebra': 'Black',
        
        # Grays / Silvers
        'granite': 'Gray', 'charcoal': 'Gray', 'graphite': 'Gray', 'slate': 'Gray', 
        'ash': 'Gray', 'agate': 'Gray', 'stone': 'Gray', 'shale': 'Gray', 
        'platinum': 'Gray', 'pewter': 'Gray', 'palladium': 'Gray', 'meteor': 'Gray',
        'flint': 'Gray', 'ocean': 'Gray',
        
        # Whites / Lights
        'chalk': 'White', 'ivory': 'White', 'pearl': 'White', 'porcelain': 'White', 
        'alabaster': 'White', 'bianco': 'White', 'magnolia': 'White', 'oyster': 'White',
        'ice': 'White', 'ceramic': 'White',
        
        # Beiges / Tans / Browns
        'parchment': 'Beige', 'linen': 'Beige', 'cream': 'Beige', 'ecru': 'Beige', 
        'luxor': 'Beige', 'cashmere': 'Beige', 'savanna': 'Beige', 'almond': 'Beige', 
        'bamboo': 'Beige', 'wheat': 'Beige', 'champagne': 'Beige', 'kalahari': 'Beige', 
        'gobi': 'Beige', 'macchiato': 'Beige', 'taupe': 'Beige', 'sand': 'Beige', 
        'dune': 'Beige', 'saddle': 'Brown', 'oak': 'Brown', 'cocoa': 'Brown', 
        'cognac': 'Brown', 'caramel': 'Brown', 'cuoio': 'Brown', 'cinnamon': 'Brown', 
        'java': 'Brown', 'havanna': 'Brown', 'havana': 'Brown', 'mocha': 'Brown', 
        'espresso': 'Brown', 'nougat': 'Brown', 'chestnut': 'Brown', 'amaro': 'Brown', 
        'sepia': 'Brown', 'truffle': 'Brown', 'walnut': 'Brown', 'tartufo': 'Brown', 
        'terra': 'Brown', 'natural': 'Brown', 'palomino': 'Tan', 'camel': 'Tan', 
        'khaki': 'Tan', 'atacama': 'Tan',
        
        # Reds / Oranges
        'salsa': 'Red', 'coral': 'Red', 'imola': 'Red', 'fox': 'Red', 
        'burgundy': 'Red', 'magma': 'Red', 'carrera': 'Red', 'maroon': 'Red', 
        'chateau': 'Red', 'bordeaux': 'Red', 'fiona': 'Red', 'scarlet': 'Red', 
        'garnet': 'Red', 'crimson': 'Red', 'ruby': 'Red', 'cabernet': 'Red', 
        'rosso': 'Red', 'sakhir': 'Orange', 'kyalami': 'Orange',
        
        # Greens / Blues
        'jade': 'Green', 'cypress': 'Green', 'forest': 'Green', 'nordkap': 'Blue', 
        'nautic': 'Blue', 'yachting': 'Blue', 'estoril': 'Blue', 'marina': 'Blue'
    }
    
    for key, val in special_map.items():
        if key in x_lower:
            return val

    # 4. Check standard baseline colors
    std_colors = [
        'black', 'white', 'gray', 'grey', 'silver', 'red', 'blue', 
        'green', 'brown', 'beige', 'yellow', 'orange', 'gold', 'purple', 'tan'
    ]
    
    for color in std_colors:
        if color in x_lower:
            return 'Gray' if color == 'grey' else color.capitalize()
            
    # 5. Everything else becomes "Other"
    return "Other"

# 1. Apply the function to create a new cleaned column
df['Cleaned_Color'] = df['Interior Color'].apply(get_main_color)

df['Cleaned_Color_2'] = df['Exterior Color'].apply(get_main_color)

# 2. Calculate the percentage of rows marked as "Other"
total_rows = len(df)
other_count = (df['Cleaned_Color'] == 'Other').sum()
percentage_other = (other_count / total_rows) * 100

print(f"Total rows: {total_rows}")
print(f"Rows marked as 'Other': {other_count}")
print(f"Percentage 'Other': {percentage_other:.2f}%\n")

# 3. Inspect the top 20 original colors that are still slipping through
print("Top 20 original colors currently mapping to 'Other':")
slipping_through = df[df['Cleaned_Color'] == 'Other']['Interior Color'].value_counts()
print(slipping_through.head(20))
```

```python
import json
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
# Load data
df = pd.read_csv("../data/frontend_data/cleaned_data_no_encoding.csv")
df['Year'] = df['auction_year'] - df['car_age']
df.head()
```

```python
# Add .reset_index() at the end
avg = df.groupby(['Make', 'Model', 'Year', 'Transmission_Type', 'Exterior Color'])['Sold_Price'].mean().reset_index()

# Now this will work perfectly
porsche = avg[(avg['Make'] == 'Porsche') & (avg['Model'] == '996 911')]
porsche
```

```python
# Use .agg() to get multiple statistics at once
stats = df.groupby(['Make', 'Model', 'Year', 'Transmission_Type', 'Exterior Color'])['Sold_Price'].agg(['mean', 'count']).reset_index()

# Rename the columns so they are easier to read
stats = stats.rename(columns={'mean': 'Average_Price', 'count': 'Total_Sold'})

# Now filter for Porsches
porsche = stats[(stats['Make'] == 'Porsche') & (stats['Model'] == '996 911')]
porsche
```
