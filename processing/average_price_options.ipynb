```python
import pandas as pd

# 1. Load historical dataset (Make sure to point to the correct CSV name if it differs)
df = pd.read_csv("../data/frontend_data/cleaned_data_no_encoding.csv")
df['Year'] = df['auction_year'] - df['car_age']

# 2. Keep only the columns we need for the fallback logic
cols_to_keep = ['Make', 'Model', 'Year', 'Transmission_Type', 'Exterior Color', 'Sold_Price']
df_light = df[cols_to_keep].dropna(subset=['Sold_Price'])

# 3. Save the lightweight raw data directly (No groupby!)
df_light.to_csv("../data/frontend_data/historical_averages_lookup.csv", index=False)
df_light.head()
```
