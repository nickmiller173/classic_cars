```python

# Builds the raw per-auction lookup used for fallback price estimates in the frontend.
# Stored as individual rows (not pre-aggregated) so the app can filter and average dynamically
# based on whatever combination of attributes the user selects.
import pandas as pd

df = pd.read_csv("../data/frontend_data/cleaned_data_no_encoding.csv")

# Derive model year from auction year minus car age (car_age = auction_year - model_year)
df['Year'] = df['auction_year'] - df['car_age']

# Minimal column set — keeps the file small while covering every attribute
# the frontend might filter on (transmission, color, drivetrain, body style).
cols_to_keep = ['URL', 'Make', 'Model', 'Mileage', 'Year', 'Transmission_Type', 'Exterior Color', 'Drivetrain', 'Body Style', 'Sold_Price']
df_light = df[cols_to_keep].dropna(subset=['Sold_Price'])

# Save individual sale rows so the app can compute averages over any filtered subset at query time.
# URL is retained to support deduplication checks on the frontend if needed.
df_light.to_csv("../data/frontend_data/historical_averages_lookup.csv", index=False)
df_light.head()
```
