```python
# Extracts a lightweight subset of cleaned auction data for use in Streamlit dashboards.
# Source is cleaned_data_no_encoding.csv which retains raw mileage and human-readable values.
import pandas as pd

df = pd.read_csv("../data/frontend_data/cleaned_data_no_encoding.csv")

# Derive model year from auction year minus car age (car_age = auction_year - model_year)
df['Year'] = (df['auction_year'] - df['car_age']).astype(int)

# Normalize seller type: collapse all dealer fee variants (e.g. "Dealer ($95 Doc Fee)") into "Dealer"
df['Seller Type'] = df['Seller Type'].apply(
    lambda x: 'Private Party' if str(x).strip() == 'Private Party' else 'Dealer'
)

# Keep only columns needed for price trend charts, feature-filter UI, and body-style/seller-type analyses.
# Binary indicator columns (2_keys_ind, etc.) drive the "extras" filter checkboxes on the dashboard.
dashboard_cols = [
    'Make', 'Model', 'Year', 'auction_year', 'auction_month', 'Sold_Price',
    'Seller Type', 'Body Style',
    '2_keys_ind', 'is_dry_climate_car', 'has_sport_seats',
    'recent_major_service', 'is_project_car', 'emissions_ind',
    'loan_ind', 'one_owner_ind', 'carfax_ind', 'owners_manual_ind',
    'has_new_tires',
]
df_dashboard = df[dashboard_cols]
df_dashboard.to_csv('../data/frontend_data/dashboard_data.csv', index=False)
```
